## Step 1. Simple clustering

The first thing that we will try is to cluster the geolocation data of the calls. The geolocation of a record consists of a pair `(latitude, longitude)`, which represents the position of the record on the map. Since the dataset that we have spans only a small geographic location, i.e. Boston, we will use the Euclidean distance when measuring how close two points are. Techinically, we should pick a [projection](https://github.com/mbostock/d3/wiki/Geo-Projections) in order to transform the data into 2d, and measure  the Euclidean distance in the projected space.

Since we are dealing with numeric data, we might consider centering and normalizing them. In order to do that, check the [StandardScaler](http://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html#sklearn.preprocessing.StandardScaler) from `scikit-learn`.

```python
geolocations = boston[['LONGITUDE','LATITUDE']]

print geolocations.head(6)
print ""
print geolocations.info()
```

###### (2) Now I will transform the data from a 3D geolocation pair `(longitude, latitude)` to a 2D pair `(x,y)`. To do this, I will use the Mercator Projection.

I will use the formulas given here, on Wolfram Alpha: http://mathworld.wolfram.com/MercatorProjection.html

Specifically, I will use the following two transformations:
\begin{align}x = \lambda - \lambda_{0} \end{align}

\begin{align}y = \frac{1}{2}\ln \left(\frac{1+sin(\phi)}{1-sin(\phi)} \right) \end{align}

We can see that the transformation of longitude $\lambda$ to $x$ is a very simple transformation. More importantly, the transformation of $\lambda$ to $x$ does not involve the latitude $\phi$. Likewise, the transformation of $\phi$ to $y$ does not involve $\lambda$.

We can write simple function that converts a geolocation pair `(longitude, latitude)` to a pair of coordinates `(x,y)`. However, because we are dealing with dataframes, it will be easier to write two separate functions, then use the `apply` feature to apply each individual function to the dataframe.

According to Google, Boston is located at 42.3601° N, 71.0589° W. Thus, I will set $\lambda_{0} = -71.0589$.

```python
from math import log, sin, pi

#Convert longitude to x
def convert_longitude(longitude):
    return (longitude - (-71.0589))
    
#Convert latitude to y
def convert_latitude(latitude):
    return 0.5*log((1+sin(latitude))/(1-sin(latitude)))
    
```

The following block of code performs the mercator projection on the longitude and latitude data

```python
print "Before: \n", geolocations.head(5)
t = pd.DataFrame.copy(geolocations) #Make a copy of the dataframe (If we don't, then 't' & 'geolocations' will have the same pointer)
t.LONGITUDE = geolocations.LONGITUDE.apply(convert_longitude)
t.LATITUDE = geolocations.LATITUDE.apply(convert_latitude)
print "\nAfter: \n", t.head(5)
```

###### (3) Next, I apply `StandardScaler` to the data to center and normalize it.

To see how the geolocation pairs compare to the mercator projection calculated above, I apply `StandardScaler` to both the geolocation data and the transformed data.

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler(copy=True, with_mean=True) 

scaled_geolocations = scaler.fit_transform(geolocations)
scaled_projection = scaler.fit_transform(t)

print type(scaled_geolocations)
print scaled_geolocations
```

###### (4) Now plot the scaled data for both the original geolocations and the projected geolocations for comparison

```python
#Set x and y for scaled geolocations
x_g = scaled_geolocations[:,0]
y_g = scaled_geolocations[:,1]

#Set x and y for scaled projection
x_p = scaled_projection[:,0]
y_p = -scaled_projection[:,1]

#Plot the data from Boston Geolocations
f, (ax1, ax2) = plt.subplots(1, 2, sharey=False)
ax1.scatter(x_g,y_g)
ax1.set_title('Scatter Plot of Geolocations in Boston')

#Plot actual map of boston
img=mpl.image.imread('neighborhoods.jpg')
ax2.imshow(img)
ax2.set_title('Map of Boston Neighborhoods')
plt.setp(ax2.get_xticklabels(), visible=False)
plt.setp(ax2.get_yticklabels(), visible=False)
ax2.grid(b=False)
f.set_size_inches(17,9)

#Plot data from scaled 
fig = plt.figure(figsize=(7.7, 9));
ax = fig.add_subplot(111);
plt.scatter(x_p,y_p);
ax.set_title('Scatter Plot of Projected Geolocations in Boston');
```

##### (5) Determine number of k-means clusters for (1) the original scaled geolocations and (2) the scaled mercator projection of the geolocations

In this section, I am going to perform k-means clustering on both datasets for k=1,2,...23.  Since the  `kmeans.inertia_` variable contains the clustering error using the L2 norm, I will consider the `kmeans.inertia_` variable at each `k` to determine the appropriate number of clusters. 

```python
import numpy as np
from sklearn.cluster import KMeans

#Calculate the error for the original geolocations data. Technically, there are 23 neighborhoods in Boston, so
#for fun, I will calculate the clustering error for k = {1, 2, ... , 23}
geo_error = np.zeros(24)
geo_error[0] = 0;
for k in range(1,24):
    kmeans = KMeans(init='k-means++', n_clusters=k, n_init=10)
    kmeans.fit_predict(scaled_geolocations)
    geo_error[k] = kmeans.inertia_
    
#Calculate the error for the mercator projected data
merc_error = np.zeros(24)
merc_error[0] = 0;
for k in range(1,24):
    kmeans = KMeans(init='k-means++', n_clusters=k, n_init=10)
    kmeans.fit_predict(scaled_projection)
    merc_error[k] = kmeans.inertia_
```

Now plot the data...

```python
#Plot the data from Boston Geolocations
f, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey=False);
f.set_size_inches(16,4);

#Geolocations error
ax1.plot(range(1,len(geo_error)),geo_error[1:])
ax1.set_xlabel('Number of clusters `k`');
ax1.set_ylabel('Error');
ax1.set_title('Geolocation Data: Number of Clusters vs. Error');

#Mercator error
ax2.plot(range(1,len(merc_error)),merc_error[1:],'r')
ax2.set_xlabel('Number of clusters `k`');
ax2.set_ylabel('Error');
ax2.set_title('Mercator Projection: Number of Clusters vs. Error');

#Overlay of geolocations & mercator
ax3.plot(range(1,len(geo_error)),geo_error[1:])
ax3.plot(range(1,len(merc_error)),merc_error[1:],'r')
ax2.set_xlabel('Number of clusters `k`');
ax3.set_ylabel('Error');
ax3.set_title('Number of Clusters vs. Error');
ax3.legend(['Geolocation','Mercator Projection'])
ax3.set_xlabel('Number of clusters `k`');
```

**1st plot**: Number of Clusters vs. Error for the scaled geolocation data

**2nd plot**: Number of Clusters vs. Error for the scaled Mercator Projection

**3rd Plot**: Plots 1 and 2 overlayed (for comparison)

___

Whether we perform clustering on the the scaled mercator projection or on the scaled geolocations, the result is essentially same, as illustrated by the third plot above.

As we see in both plots, the clustering error decreases exponentially as the number of clusters $k$ increases. As $k$ approaches 5, the rate of error reduction begins to decrease significantly. We can thus say that $k=5$ is the recommended number clusters for this data, as $k=5$ is roughly the "elbow" of both graphs.

###### (6) Cluster both datasets (`scaled_geolocations` and `scaled_projection`) with k=5, then plot

```python
#Initialize kmeans objects
kmeans_geo = KMeans(init='k-means++', n_clusters=5, n_init=10)
kmeans_proj = KMeans(init='k-means++', n_clusters=5, n_init=10)

#Fit and predict
labels_geo = kmeans_geo.fit_predict(scaled_geolocations)
labels_proj = kmeans_proj.fit_predict(scaled_projection)

#Plot
f, (ax1, ax2) = plt.subplots(1, 2, sharey=False)
f.set_size_inches(17,9)
colors = ['#42D0FF', '#FFB53D', '#FF3DAE','#993366','#00CC66']

for k in range(0,5):
    cluster = labels_geo == k
    ax1.scatter(scaled_geolocations[cluster, 0], scaled_geolocations[cluster, 1], color=colors[k], edgecolor='k')
ax1.set_title('Clustered Geolocations Data (k=4 clusters)')
ax1.set_xlabel('Normalized Longitude')
ax1.set_ylabel('Normalized Latitude')

for k in range(0,5):
    cluster = labels_proj == k
    ax2.scatter(scaled_projection[cluster, 0], -scaled_projection[cluster, 1], color=colors[k], edgecolor='k')
ax2.set_title('Clustered Projected Data (k=4 clusters)')
ax2.set_xlabel('Normalized x')
ax2.set_ylabel('Normalized y')

plt.show()
```

The above two plots illustrate the difference in k-means clustering between the geolocation data and the projected geolocation data. Although we do expect to see a difference in clustering between the two datasets, it is interesting to see how applying the Mercator Projection to the data changes the way k-means clusters the data.

Realistically, the clustering above isn't very meaningful. Because we are only clustering the 311 call geolocations, we are only obtaining information about the neighborhoods of Boston. However, for each 311 call, we _already_ have its geolocation and the neighborhood that geolocation is located in. Thus, this clustering does not provide us with very useful information. In fact, if we want to be technical, we should in theory expect 23 different clusters, one for each neighborhood.

###### (3) Since we can't input these columns directly into a k-means clustering algorithm, we're going to have to convert these strings to numbers

Here, I encode the `SUBJECT` and `Source` columns (i.e., I convert each string in those columns to a numerical value). I then round the `LONGITUDE` and `LATITUDE` columns to the nearest 10ths place.

In general, I am interested in seeing how `SUBJECT` and `Source` relate to geolocation. I round the geolocation variables because I just want to consider *general* areas of Boston. For example, if there are a lot of people near Fenway Park reporting 311 cases, I want to consider all of those 311 calls as coming from the same location. I don't want to consider each one of those Fenway calls as individual calls. Rounding the data in this way will allow for "cleaner" clustering.

```python
#Import LabelEncoder. This function converts strings to labels (integers)
from sklearn.preprocessing import LabelEncoder

updated_data = pd.DataFrame.copy(new_data)

#Function to encode
def encode(x):
    LE = LabelEncoder()
    labels = LE.fit_transform(x) + 0.0 #Convert labels from int64 to float by adding 0.0
    return labels

#Now encode
updated_data.SUBJECT = encode(new_data.SUBJECT)
updated_data.Source = encode(new_data.Source)
updated_data.LONGITUDE = np.round(new_data.LONGITUDE,1) #Round the longitude to the nearest 10ths place
updated_data.LATITUDE = np.round(new_data.LATITUDE,1) #Round the latitude to the nearest 10ths place

#Print out new data
updated_data.head(5)
```

###### (4) Now we can apply the standard scaler to this data 

```python
ss = StandardScaler(copy=True, with_mean=True) 

scaled_updated_data = ss.fit_transform(updated_data)
print scaled_updated_data.shape
print ""
print scaled_updated_data
print ""
print np.unique(scaled_updated_data)
```

###### (5) As before, we want to perform k-means, but first determine the appropriate number of clusters

The ideal number of clusters `k` is marked with a red star above.

Now perform k-means using `k=7` clusters:

```python
k=7
kmeans = KMeans(init='k-means++', n_clusters=k, n_init=10)
labels = kmeans.fit_predict(scaled_updated_data)
idx = np.argsort(labels)
rX = scaled_updated_data[idx,:]


f, (ax1, ax2) = plt.subplots(1, 2, sharey=False);
f.set_size_inches(12,5);

#Use Matplotlib's 'pcolor'. (It's basically the same thing as seaborn's heatmap, but it runs faster than seaborn... and it doesn't crash)
ax1.pcolor(scaled_updated_data,cmap=plt.cm.Paired);
ax2.pcolor(rX,cmap=plt.cm.Paired);

#Modify axes
ax1.axis('tight');
ax1.axis('off');
ax2.axis('tight');
ax2.axis('off');

#Set titles now...
ax1.set_title('Data Set Before Clustering');
ax2.set_title('Data Set After Clustering');
plt.show();
```

Generally speaking, any value greater than `mean +/- 2 * std` is an outlier. We will calculate outliers this way:

```python
#First, separate by clusters
C1 = euclidean_dists[euclidean_dists['Label']==0]
C2 = euclidean_dists[euclidean_dists['Label']==1]
C3 = euclidean_dists[euclidean_dists['Label']==2]
C4 = euclidean_dists[euclidean_dists['Label']==3]
C5 = euclidean_dists[euclidean_dists['Label']==4]
C6 = euclidean_dists[euclidean_dists['Label']==5]
C7 = euclidean_dists[euclidean_dists['Label']==6]

#Now find means
C1_mean = C1.Euc_Dist.mean(axis=0)
C2_mean = C2.Euc_Dist.mean(axis=0)
C3_mean = C3.Euc_Dist.mean(axis=0)
C4_mean = C4.Euc_Dist.mean(axis=0)
C5_mean = C5.Euc_Dist.mean(axis=0)
C6_mean = C6.Euc_Dist.mean(axis=0)
C7_mean = C7.Euc_Dist.mean(axis=0)

#...and standard deviations
C1_std = C1.Euc_Dist.std(axis=0)
C2_std = C2.Euc_Dist.std(axis=0)
C3_std = C3.Euc_Dist.std(axis=0)
C4_std = C4.Euc_Dist.std(axis=0)
C5_std = C5.Euc_Dist.std(axis=0)
C6_std = C6.Euc_Dist.std(axis=0)
C7_std = C7.Euc_Dist.std(axis=0)

#Store in numpy array
means = np.array([C1_mean,C2_mean,C3_mean,C4_mean,C5_mean,C6_mean,C7_mean])
stdevs = np.array([C1_std,C2_std,C3_std,C4_std,C5_std,C6_std,C7_std])

#This stores the thresholds (mean + 2*stdev)
thresholds = means+2*stdevs

print thresholds
```
