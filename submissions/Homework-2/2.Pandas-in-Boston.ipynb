Great! Now, let's check all the different values of the column `CASE_TITLE` together with their counts **(5 pts)**

````python
# Write code that prints the different values of the column 'CASE_TITLE'
# and the count for the occurencies of each value in the boston dataframe

#All of the cases and their number of occurrences will be stored in a dictionary called 'cases'
cases = dict()

#This for loop iterates through all the rows in 'boston'. For each row, it grabs the case title... then 
#it checks if the case title is already in the dictionary. If the case title IS in the dictionary, it
#increments that case's count. Else, the title gets added to the dictionary and its count is set to 1.
for idx, row in boston.iterrows():
    case_title = row['CASE_TITLE']
    if (case_title in cases):
        cases[case_title]+=1
    else:
        cases[case_title] = 1

#Convert 'cases' to a dataframe
cases_df = pd.DataFrame(cases.items(), columns=['Case', 'Count']) #Converts the dictionary from part 2 into a dataframe
cases_df = cases_df.sort(['Case']) #Sort the dataframe by case title
cases_df = cases_df.reset_index(drop=True)
print cases_df
````

Now, let's plot (preferably with horizontal bars) the counts for all the `CASE_TITLE`s that occur more than 250 times in the dataset. **(5 pts)**

````python
# Plots the high-occuring categories for cases as a horizontal histogram.

#The following loop extracts the high-occurring cases from 'cases', a dictionary i created earlier which contains 
#case titles and how many times each case title occurred
high_occurrences = dict()
for key in cases:
    if (cases[key] > 250):
        high_occurrences[key] = cases[key]

#Initialize figure
f = plt.figure(figsize=(7,7))
ax = f.add_subplot(111)
ax.yaxis.set_label_position("right") #y-axis label will be placed on the right hand side of the graph instead of the left

#Plot figure, titles, axis labels, etc.
x = range(len(high_occurrences))
data = plt.barh(x,high_occurrences.values(),align='center',color=[0.2, 0.6, 0.5])
plt.yticks(x,high_occurrences.keys());
plt.xlabel('Number of Occurrences');
plt.ylabel('Case Title',rotation=90);
plt.title('January & February 2015: Case titles which occur more than 250 times',fontweight='bold');


#This just plots the actual # of occurrences next to each bar
y=0
for key in high_occurrences:
    value = high_occurrences[key]
    ax.text(value+100,y-0.35,'%d'%value);
    y+=1
````

As you can see, the most common categories are related to snow. Let us see which neighborhoods were mostly affected by snow problems. This information is saved under the `neighborhood` column. Make a plot (similar to the one above, like a horizontal histogram) that for each neighborhood shows the number of snow-related calls made from there. **(5 pts)**

````python
#Snow data will be stored in this dictionary.
snow = dict()

#The following loop finds all rows in the 'boston' dataframe whose "CASE_TITLE" column contains the word 'snow'
for idx, row in boston.iterrows():
    try:
        case_title = row['CASE_TITLE'].lower()
        if ('snow' in case_title):
            neighborhood = row['neighborhood']
            if (neighborhood in snow):
                snow[neighborhood] +=1
            else:
                snow[neighborhood] = 1
    except (AttributeError):
        continue

#Initialize figure
f = plt.figure(figsize=(7,7))
ax = f.add_subplot(111)
ax.yaxis.set_label_position("right") #y-axis label will be placed on the right hand side of the graph instead of the left

#Plot figure, titles, axis labels, etc. (NOTE: I am purposely plotting neighborhoods labeled as 'nan')
x = range(len(snow))
data = plt.barh(x,snow.values(),align='center',color=[0.2, 0.6, 0.5])
plt.yticks(x,snow.keys());
plt.xlabel('Number of Snow-Related Calls');
plt.ylabel('Neighborhood',rotation=90);
plt.title('Number of Snow-Related Calls by Neighborhood (January and February 2015)',fontweight='bold');


#This just plots the actual # of occurrences next to each bar
y=0
for key in snow:
    value = snow[key]
    ax.text(value+50,y-0.35,'%d'%value);
    y+=1
````

## Step 3. Advanced filtering & joins

By now, you should have noticed that, for each day, there are multiple records in our dataframe. Use the `groupby` function to get, for each day, the counts for all the case categories. Plot the result for the top 5 categories as you see fit. **(20 pts)**

````python
# Write your code here

'''
    EXPLANATION OF MY CODE:

    In step 2, I created a dictionary called 'high_occurrences'. This dictionary contains all of the case titles which
    occur more than 250 times in the dataset.
    
    I call 'high_occurrences' here and convert it to a Pandas dataframe called 'd'. This dataframe has 2 columns: 'Case'
    and 'Count'. 'Case' is short for 'CASE_TITLE' and 'Count' represents the number of times a case title occurrs. With
    this newly-created dataframe, I am able to figure out the top 5 reported cases by sorting the dataframe by 'Count'. 
    
    The top 5 reported case titles are then saved as a list called 'titles'. I then use this list to establish a new
    dataframe, which I call 'top5'. This dataframe becomes the final dataframe used for plotting the data.
    
    Using the 'groupby' function, I am able to iterate through the original 'boston' dataset by date. I set the date
    frequency to '1d' so that I iterate through the dataset by each day, not by each timestamp.  
    
    Within this iteration loop, I create another loop, which iterates through all the case titles for the current date.
    I take the length of the dataframe 'case_data' in order to grab the total number of cases for that particular case
    title. I then append the total number of cases to the appropriate index of 'temp_list'. After I am finished iterating
    through all of the case titles for the current date, I add 'temp_list' as a new row to the dataframe 'top5'.
    
    Once I am finished iterating through the entire dataset by date, I plot the data. There are 2 plots:
    (1) Just a normal lineplot of the data
    (2) Stacked lineplot so you can see the *total* number of 'top 5' cases per day, broken down by case type

'''


#Get highest number of occurrences
d = pd.DataFrame(high_occurrences.items(), columns=['Case', 'Count']) #Converts the dictionary from part 2 into a dataframe
d = d.sort(['Count'],ascending=0) #Sort the dataframe by count
d = d.head(5) #take the top 5 cases

#These are the case titles we care about
titles = d.Case.tolist() 

#Create dataframe
top5 = pd.DataFrame(columns=titles)

#Iterate through 'boston' by date (e.g., Jan 1st, Jan 2nd, Jan 3rd, Jan 4th...)
for time, data in boston.groupby(pd.Grouper(freq='1d',key='OPEN_DT')):


    temp_list = [0,0,0,0,0]
    
    #For each date, iterate through all the CASE_TITLE groups:
    for case_title, case_data in data.groupby('CASE_TITLE'):
        
        #get number of cases for the current case title
        num_cases = len(case_data)
        
        #Append to the correct index of 'temp_list'
        if (case_title == titles[0]):
            temp_list[0] = num_cases
        elif (case_title == titles[1]):
            temp_list[1] = num_cases
        elif (case_title == titles[2]):
            temp_list[2] = num_cases
        elif (case_title == titles[3]):
            temp_list[3] = num_cases
        elif (case_title == titles[4]):
            temp_list[4] = num_cases

    #Save row to dataframe
    top5.loc[time] = temp_list
    
f = plt.figure();
fig, axes = plt.subplots(nrows=2,figsize=(11,7), sharex=True);

top5.plot(ax=axes[0],colormap='Set2');
top5.plot(stacked=True, kind='area',ax=axes[1],colormap='Set2');

fig.suptitle('\'Top 5\' Case Categories for January and February 2015', fontsize=14,fontweight='bold');
axes[0].set_ylabel('# of Cases');
axes[1].set_ylabel('Total # of Cases');
axes[1].set_xlabel('Date');
axes[0].set_title('Top 5 Case Categories by Day');
axes[1].set_title('Total Number of \'Top 5\' Case Categories per Day (Individual Contributions Shown)');

axes[0].grid(b=True, which='minor', color=[0.65, 0.65, 0.65], linestyle=':');
axes[0].grid(b=True, which='major', color=[0.65, 0.65, 0.65], linestyle=':');
axes[1].grid(b=True, which='minor', color=[0.65, 0.65, 0.65], linestyle=':');
axes[1].grid(b=True, which='major', color=[0.65, 0.65, 0.65], linestyle=':');
    
````

Let's see how this trend compares to the snowfall. First, we need to read the data from `snow.csv`

````python
# Read the dataframe
snow_data = pd.read_csv("snow.csv",parse_dates=[1])
print snow_data
````

According to the documentation, the column `SNOW` has the values for the snowfall, in mm. However, the dataset we have has information from many different stations around Boston. Aggregate all their measurements and transform the result to centimeters. Finally, plot the result **(10 pts)**

````python
# Write your code here

#First, clean the data. The values in the column 'SNOW' should be greater than or equal to  0.
cleaned_snow_data = snow_data[snow_data.SNOW >= 0]

#Now reorganize by date, as before
reorg_snow = cleaned_snow_data.groupby([pd.Grouper(freq='1d',key='DATE')])

#Aggregate by mean
reorg_snow = reorg_snow.agg('mean')

#Delete 'SNWD' column now
reorg_snow = reorg_snow.drop('SNWD', axis=1)

#Convert mm to cm by diving the whole 'SNOW' column by 10
reorg_snow.SNOW = reorg_snow.SNOW / 10

#Replace all nan's with 0s
reorg_snow = reorg_snow.fillna(0)

#Initialize & plot figure
f = plt.figure(figsize=(10,3))
ax = f.add_subplot(111)
reorg_snow.plot(ax=ax,color=[0.2, 0.6, 0.8])
ax.set_ylabel('Snowfall (cm)')
ax.set_xlabel('Date')
ax.set_title('Aggregated Snow Measurements for January & February 2015',fontsize=14,fontweight='bold')
ax.patch.set_facecolor('white')
ax.grid(b=True, which='minor', color=[0.8, 0.8, 0.8], linestyle=':')
ax.grid(b=True, which='major', color=[0.8, 0.8, 0.8], linestyle=':')
````
